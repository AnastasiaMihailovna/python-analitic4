## Задача 3 Составьте несколько сводных таблиц

### Задача 3.1 Найдите среднюю стоимость домов в зависимости от количества спален

```python
df.groupby ('bedrooms').agg({'price':'mean'}).sort_values('price')
```

### Задача 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby ('delta_renovated').agg ({'price':['min','max','mean']})
```

### Задача 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table (index='waterfront',columns='view', aggfunc='count',fill_value=0)
```

### Задача 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
df.pivot_table (index='floors',columns='bedrooms', aggfunc='count',fill_value=0)
```

Ответ: Больше всего 2х этажных домов с 4 спальнями

```python
pd.crosstab(index=df ['floors'],columns=df ['bedrooms'])
```

### Задача 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df.groupby(['delta_renovated', 'yr_renovated']).agg({'price':'median'}).unstack('delta_renovated').fillna(0)
```
